```python
# todo: switch to threaded
def api_wrapper(url):
    result = requests.get(
        "https://oed-api.oxforddictionaries.com" + url,
        headers={"app_key":"e924ba24525f6c231fbfeba62ce965ca", "app_id": "b0691462"} # Not sure where these came from. Perhaps auto-generated?
    )
    result.raise_for_status()
    return json.loads(result.text)
    
def get_next(first=0,bound=100):
    return api_wrapper(
        f'/oed/api/v0.1/senses/?limit={bound}&offset={first}', # then offset = 100, 200, etc
    )
```

```python
def next_link(json_arr):
    ind, _ = max(enumerate(map(lambda x: int(parse_qs(urlparse(x["links"]["next"]).query)['offset'][0]), json_arr)), key=itemgetter(1))
    return json_arr[ind]["links"]["next"] 

# Simpler way to do this: function that takes array of json and gives me the next link to use
# That way we don't care if it was loaded from disk or is in memory

# Downside: can't thread it
```

```python
# bootstrapping
# cache = [get_next(0,100)]
with open('oed.json', 'r') as infile:
    cache = json.load(infile)
```

```python
for i in range(0,8000):
    try:
        print(i)
        cache.append(api_wrapper(next_link(cache)))
        print(cache[-1]['data'][-1]['lemma'])
        print(next_link(cache))
        print(datetime.now())
        if(i%100 == 0):
            with open('oed.json', 'w') as outfile: # consider writing to tar gz straight away to save dropbox space (python crashes)
                json.dump(cache, outfile)
    except requests.HTTPError as e:
        if (e.response.status_code == 504):
            sleep(300)
            print("504, holding off")
            print(datetime.now())
            continue
        else:
            raise
```
